```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
#Generating dublicate image from imported image throught Augmentation
img_gen = ImageDataGenerator(
    rotation_range = 40,
    width_shift_range= 0.2,
    height_shift_range= 0.2,
    shear_range= 0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
#Load the image 
img = load_img("/content/Cat_November_2010-1a.jpg")
img
```

```python
#Converting the image into array
x = img_to_array(img)
#Rechaping the image array with 1,3,150,150
x = x.reshape((1,)+x.shape)
```

```python
#Generating the image and save on preview
i=0
for batch in img_gen.flow(x,batch_size=1,
                          save_to_dir='/home', save_prefix='Dat', save_format='jpeg'):
  i +=1
  if i >20:
    break
```
